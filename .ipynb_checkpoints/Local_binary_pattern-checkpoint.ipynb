# Local Binary Pattern

<p>Dalam mengekstraksi ciri citra tentunya banyak cara yang bisa digunakan, salah satunya adalah menggunakan algoritma Local Binary Pattern (LBP) yang akan dibahas.</p>
    
   <p>Local Binary Pattern (LBP) adalah operator tekstur sederhana namun sangat efisien yang melabeli piksel gambar dengan membatasi lingkungan setiap piksel dan menganggap hasilnya sebagai bilangan biner. Ide dasar untuk mengembangkan operator LBP adalah bahwa tekstur permukaan dimensi dapat dijelaskan dengan dua ukuran yang saling melengkapi: pola spasial lokal dan kontrak skala abu-abu.</p>
   <p>Operator LBP pertama kali dikenalkan oleh Ojala et al, dan telah terbukti sebagai deksriptor tekstur yang tangguh. Operator LBP melabeli piksel-piksel dari sebuah citra dengan melakukan proses thresholding ketetanggaan 3x3 dari masing-masing piksel sebagai nilai tengah dan mengubah hasilnya menjadi nilai biner, dan 256-bin LBP histogram digunakan sebagai texture descriptor. Bilangan biner yang dihasilkan (disebut Local Binary Pattern atau LBP codes) mengkodekan lokal primitif termasuk variasi dari lengkungan sisi, bintik, area datar, dsb.</p>
   <p>Bebrapa penelitian dengan metode LBP sudah diimplementasikan dalam mendukung klasifikasi berdasarkan ciri tekstur seperti untuk pengenalan wajah, sidik jadi, telapak tangan dan pemilik tanda tangan. Secara umum algoritma ini mempunyai 2 langkah:</p>

<ul>
    <li>Langkah thresholding</li>
    <li>Langkah encoding</li>
</ul>
<p>
    Pada langkah thresholding, semua nilai pixel tetangga pada setiap pola kaan dikomprasi dengan nilai yang ada dipusat diantara nilai-nilai pola mereka, ini untuk dikonversi nilai mereka mejadi nilai biner yaitu (0 dan 1). Untuk nilai 0 yang mempunyai nilai lebih kecil dengan nilai tengah atau pusat, sedangkan nilai 1 untuk angka lebih besar dari nilai pusatnya. Kemudian dalam langkah encoding, sejumlah nilai biner yang diperoleh dari langkah thresholding akan dirubah atau dikonversi kedalam bilangan  desimal untuk karakteristikdari struktur pola. Tahapan proses threshold dan encoding terdapat pada gambar dibawah.
</p>

<h3> Import Libary </h3>

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# import file citra
gambar = cv2.imread("./image/lena.png")

jml_baris = len(gambar) 
jml_kolom = len(gambar[0])

# bgr to grayscale
gambar_gray = cv2.cvtColor(gambar, cv2.COLOR_BGR2GRAY)

# buat matrix baru untuk menampung citra yang akan dikonvolusi
matriks_baru = np.zeros((jml_baris, jml_kolom))
```

<p>Untuk mendapatkan nilai LBP, kode-kode biner yang telah didapatkan, diklikkan dengan pembobotan binernya. Pada gamabr diatas menunjukkan operasi dasar LBP. Pola-pola biner LBP merepresentasikan bemacam-macam pola tepi, titik, flat areas, dan sebagainya. Unutk gamabran perhotungan dari operasi dasar LBP terdapat pada gambar dibawah.</p>

<img src="https://binus.ac.id/bandung/wp-content/uploads/2020/12/Capture2-1.jpg" height=150 width=450>

<h3> Membuat fungsi untuk proses LBP </h3>

```python
# buat fungsi untuk mengkonversi biner to desimal
def binaryToDecimal(n):
    return int(n,2)

# buat fungsi metode LBP proses encode
def lbp_method(image, nilai_tengah):
    if image >= nilai_tengah:
        image = 1
    else:
        image = 0
    return image
```

<p>Teknik dasar pada local binary pattern (LBP) adalah melakukan analisis ketetanggaan terhadap setiap pixel pada sebuah citra, sebagai contoh sebuah gambar direpresentasikan oleh matriks (Bianconi & Fernández, 2014). Namun dalam proses pixel perpixel dengan menggunakan metode local binary pattern dapat dilihat pada Gambar di bawah.</p>

<img src="https://binus.ac.id/bandung/wp-content/uploads/2020/12/Capture4.jpg" height=150 width=450>

<h3>Proses Konvolusi</h3>

```python
for brs in range(1, jml_baris - 1):
    for klm in range(1, jml_kolom - 1):
        e = gambar_gray[brs, klm]      #tengah matriks
        a = lbp_method(gambar_gray[brs-1,klm-1], e)   #kiri atas
        b = lbp_method(gambar_gray[brs-1,klm], e)     #tengah atas
        c = lbp_method(gambar_gray[brs-1, klm+1], e)   #kanan atas
        d = lbp_method(gambar_gray[brs, klm-1], e)     #kiri
        f = lbp_method(gambar_gray[brs, klm+1], e)     #kanan
        g = lbp_method(gambar_gray[brs+1, klm-1], e)   #kiri bawah
        h = lbp_method(gambar_gray[brs+1, klm], e)     #tengah bawah
        i = lbp_method(gambar_gray[brs+1, klm+1], e)   #kanan bawah

        #hitung hasil konvolusi
        biner = f"{f}{c}{b}{a}{d}{g}{h}{i}"
        konvolusi = binaryToDecimal(biner)
        # print(konvolusi)

        #isikan hasil konvolusi ke matriks_baru
        matriks_baru[brs, klm] = konvolusi

matriks_baru = matriks_baru.astype(np.uint8)
```

<p>Pada gambar diatas mengambarkan proses ekstraksi ciri citra dengan ukuran window 3×3 pixel. Setelah didapatkan nilai desimal dari setiap perhitungan biner pada semua pixel, proses terakhir adalah membuatkan histogram dari semua nilai desimal yang telah dihasilkan sebelumnya.</p>

```python
plt.title("original")
plt.imshow(gambar_gray, cmap="gray")
plt.show()
```

```python
plt.title('local binary pattern')
plt.imshow(matriks_baru, cmap="gray")
plt.show()
```

<p>Operator LBP hanya terdiri dari beberapa piksel tetangga dengan operasi perhitungan yang tidak rumit. Secara umum gambaran dari kerangka kerja dari local binary pattern dan metode yang terkait dalam mendukung  operator LBP dapat dijabarkan pada Gambar dibawah.</p>

<img src="https://binus.ac.id/bandung/wp-content/uploads/2020/12/Capture3.jpg" height=350 width=450>

<h3> Proses Histogram </h3>

```python
gray_value = np.zeros(256)

for x in range(len(matriks_baru)):
	for i in range(len(matriks_baru[0])):
		pixel = matriks_baru[x, i]
		gray_value[pixel] += 1
```

```python
a = np.array(gray_value)
fig, ax = plt.subplots(figsize =(5, 3))
ax.hist(a, bins = [0, 25, 50, 75, 100, 125, 150, 175, 200, 225, 255])
plt.show()
```
